# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import cm 
```

# Function

# $$ f(x_1,x_2) =5x_1^2+x_2^2+4x_1x_2^2-14x_1-6x_2+14 $$ 

# Convert then mathematical function into python function

```python
def f(x_1,x_2):
    
    return 5*x_1**2 + x_2**2 + 4*x_1*x_2 - 14*x_1 - 6*x_2 - 14
```

# Creat then some x_1's and x_2's.....Data

```python
x_1 = np.linspace(start=-3, stop=3, num=200)
x_2 = np.linspace(start=-3, stop=3, num=200)
x_1, x_1 = np.meshgrid(x_1, x_2)
```

# Plot the function

```python
fig=plt.figure(figsize=(20, 18))
ax=fig.gca(projection = '3d')
ax.set_xlabel('X_1', fontsize=20)
ax.set_ylabel('X_2', fontsize=20)
ax.set_zlabel('f(X_1, X_2) -- Cost', fontsize=20)
ax.azim = 120
ax.plot_surface(x_1, x_2, f(x_1, x_2), cmap = cm.winter, alpha = 0.4)
plt.show()
```

# Partial Derivatives of the given function

# $$ \frac{\partial f}{\partial x_1} = {10x_1 + 4x_2 -14}$$

# $$ \frac{\partial f}{\partial x_2} = {2x_2 + 4x_1 -6}$$

# Convert the above Partial derivatives into python function.

```python
def fpx_1(x_1, x_2):
    
    return 10*x_1 + 4*x_2 -14
```

```python
def fpx_2(x_1, x_2):
    
    return 2*x_2 + 4*x_1 -6
```

# Gradient Descent

```python
params = np.array([2.8,  2.8])

max_iter = 5000

multiplier = 0.01

values_array = params.reshape(1,2)

for n in range(max_iter):
    
    gradient_x_1 = fpx_1(params[0], params[1])
    
    gradient_x_2 = fpx_2(params[0], params[1])
    
    gradients = np.array([gradient_x_1, gradient_x_2])
          
    params = params - multiplier *  gradients
    
    values_array = np.append(values_array, params.reshape(1,2), axis =0)
```
