# Gradient using a python function

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_1 = np.linspace(start = -4, stop = 3, num = 300)
```

```python
def f(x):
    return x**2 + x + 1
```

```python
def df(x):
    
    return 2*x + 1
```

```python
def gradient_descent(derivative_func, initial_guess, multiplier, max_iter):
    
    
    new_x = initial_guess
    
    x_list = [new_x]
    
    slope_list = [derivative_func(new_x)]
    
    for i in range(max_iter):
        
        
        previous_x = new_x
        
        gradient = derivative_func(previous_x)
        
        new_x = previous_x - multiplier * gradient
        
        x_list.append(new_x)
        
        slope_list.append(derivative_func(new_x))
        
    return x_list, slope_list, new_x
```

```python
gradient_descent(derivative_func=df,initial_guess=3,multiplier=0.1, max_iter=50 )
```

```python
plt.figure(figsize=(18,8))

plt.subplot(1,2,1)

plt.title('GD PLOT ON A SIMPLE COST FUNCTION', fontsize = 15, color = 'navy')
plt.xlabel('X_1', fontsize = 10, color = 'red')
plt.ylabel('f(X_1)', fontsize = 10, color = 'red')
plt.grid()
plt.plot(x_1, f(x_1))
plt.scatter(x_list, f(np.array(x_list)), color = 'red')


plt.subplot(1,2,2)
plt.title('GD PLOT ON THE DERIVATIVE OF THE SIMPLE COST FUNCTION', fontsize = 15, color = 'navy')
plt.xlabel('X_1', fontsize = 10, color = 'red')
plt.ylabel('df(X_1)', fontsize = 10, color = 'red')
plt.grid()
plt.plot(x_1, df(x_1))
plt.scatter(x_list, slope_list, color = 'red')

plt.show()
```
