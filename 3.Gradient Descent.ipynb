# $$g(x)=x^4-48x^2+5$$

```python
# 20 minutes to plot...
# send me screen shots....into my channel ...not on the class channel
```

```python
x_2 = np.linspace(-3, 3, 1000)
```

```python
def g(x):
    return x**4 - 4*x**2 + 5
```

```python
def dg(x):
    return 4*x**3 -8*x
```

```python
plt.figure(figsize=(18,8))

plt.subplot(1,2,1)

plt.title('PLOT OF THE SECOND COST FUNCTION', fontsize = 15, color = 'navy')
plt.xlabel('X_2', fontsize = 10, color = 'red')
plt.ylabel('g(X_2)', fontsize = 10, color = 'red')
plt.grid()
plt.xlim(-2,2)
plt.ylim(0.5,6)
plt.plot(x_2, g(x_2), linewidth = 3)



plt.subplot(1,2,2)
plt.title('THE DERIVATIVE OF THE SECOND COST FUNCTION', fontsize = 15, color = 'navy')
plt.xlabel('X_2', fontsize = 10, color = 'red')
plt.ylabel('dg(X_2)', fontsize = 10, color = 'red')
plt.grid()
plt.xlim(-2,2)
plt.ylim(-6, 8)
plt.plot(x_2, dg(x_2), linewidth = 3, color = 'navy')


plt.show()
```

```python
new_x = 2

previous_x = 2

gamma = 0.01 # also called learning_rate or multiplier.....TBD...later

x_list = [new_x]

slope_list = [dg(new_x)]


for n in range(200):
    
    previous_x = new_x
    
    new_x = previous_x - gamma*dg(previous_x)
    
    x_list.append(new_x)
    
    slope_list.append(dg(new_x))
```

```python
plt.figure(figsize=(18,8))

plt.subplot(1,2,1)

plt.title('GD PLOT ON THE SECOND COST FUNCTION', fontsize = 15, color = 'navy')
plt.xlabel('X_2', fontsize = 10, color = 'red')
plt.ylabel('g(X_2)', fontsize = 10, color = 'red')
plt.grid()
plt.xlim(-2,2)
plt.ylim(0.5,6)
plt.plot(x_2, g(x_2), linewidth = 3)
plt.scatter(x_list, g(np.array(x_list)), color = 'red', s = 100)


plt.subplot(1,2,2)
plt.title('GO ON THE DERIVATIVE OF THE SECOND COST FUNCTION', fontsize = 15, color = 'navy')
plt.xlabel('X_2', fontsize = 10, color = 'red')
plt.ylabel('dg(X_2)', fontsize = 10, color = 'red')
plt.grid()
plt.xlim(-2,2)
plt.ylim(-6, 8)
plt.plot(x_2, dg(x_2), linewidth = 3, color = 'navy')
plt.scatter(x_list, slope_list, color = 'red', s = 100)

plt.show()
```

# $$h(x)=x^5-2x^4+2$$

```python
# use linspace, from -2.5 to 2.5 with 1000
# multiplier = 0.02 same as gamma also called learning_rate
# plot the function and the slope of the function...dont do the gradient descent....
```

```python
x_3 = np.linspace(-2.5, 2.5, 1000)
```

```python
def h(x):
    
    return x**5 - 2*x**4 + 2 
```

```python
def dh(x):
    
    
    return 5*x**4 - 8*x**3
```

```python
plt.figure(figsize=(18,8))

plt.subplot(1,2,1)

plt.title('PLOT OF THE THIRD COST FUNCTION', fontsize = 15, color = 'navy')
plt.xlabel('X_3', fontsize = 10, color = 'red')
plt.ylabel('h(X_3)', fontsize = 10, color = 'red')
plt.grid()
plt.xlim(-1.2,2.5)
plt.ylim(-1,4)
plt.plot(x_3, h(x_3), linewidth = 3)



plt.subplot(1,2,2)
plt.title('THE DERIVATIVE OF THE THIRD COST FUNCTION', fontsize = 15, color = 'navy')
plt.xlabel('X_3', fontsize = 10, color = 'red')
plt.ylabel('dh(X_3)', fontsize = 10, color = 'red')
plt.grid()
plt.xlim(-1,2)
plt.ylim(-4, 5)
plt.plot(x_3, dh(x_3), linewidth = 3, color = 'navy')


plt.show()
```

```python
new_x = 0.2

previous_x = 0.2

gamma = 0.02 # also called learning_rate or multiplier.....TBD...later

x_list = [new_x]

slope_list = [dh(new_x)]


for n in range(200):
    
    previous_x = new_x
    
    new_x = previous_x - gamma*dh(previous_x)
    
    x_list.append(new_x)
    
    slope_list.append(dh(new_x))
```

```python
plt.figure(figsize=(18,8))

plt.subplot(1,2,1)

plt.title('PLOT OF THE THIRD COST FUNCTION', fontsize = 15, color = 'navy')
plt.xlabel('X_3', fontsize = 10, color = 'red')
plt.ylabel('h(X_3)', fontsize = 10, color = 'red')
plt.grid()
plt.xlim(-1.2,2.5)
plt.ylim(-1,4)
plt.plot(x_3, h(x_3), linewidth = 3)
plt.scatter(x_list, h(np.array(x_list)), color = 'red', s = 100)



plt.subplot(1,2,2)
plt.title('THE DERIVATIVE OF THE THIRD COST FUNCTION', fontsize = 15, color = 'navy')
plt.xlabel('X_3', fontsize = 10, color = 'red')
plt.ylabel('dh(X_3)', fontsize = 10, color = 'red')
plt.grid()
plt.xlim(-1,2)
plt.ylim(-4, 5)
plt.plot(x_3, dh(x_3), linewidth = 3, color = 'navy')
plt.scatter(x_list, slope_list, color = 'red', s = 100)

plt.show()
```
